# Object-oriented interface

While it is easy to quickly generate plots with the matplotlib.pyplot module, the use of object-oriented approach is recommended as it gives more control and customization of your plots. Most of the functions are also available in the matplotlib.axes.Axes class.

The main idea behind using the more formal object-oriented method is to create figure objects and then just call methods or attributes off of that object. This approach helps better in dealing with a canvas that has multiple plots on it.

In object-oriented interface, Pyplot is used only for a few functions such as figure creation, and the user explicitly creates and keeps track of the figure and axes objects. At this level, the user uses Pyplot to create figures, and through those figures, one or more axes objects can be created. These axes objects are then used for most plotting actions.

To begin with, we create a figure instance which provides an empty canvas.

## How to make ax

	注意，一般都有两种方法建立axes
	fig = plt.figure()
	
	ax = fig.add_axes([x0, y0, width, height]) # method-1, create absolute axes, note, this will lead to issues such as not enough space for ticks, all the numbers are between [0, 1].
	
	ax = fig.add_subplot(rcn) # r--> row number, c --> column number, n --> current plot position
	
	ax = fig.add_subplot(111) # do you think this is equivalent to add_axes([0,0,1,1])? No, this is actually equal to [0.125,0.11,0.775,0.77], some spaces are reserved for ticks, labels, etc.
	
	The preferred method is:
	
	fig, ax = plt.subplots(1, 3) # ax will be an array, to access them, use ax[0], ax[1], ax[2]; if we have a 2d array, then we need do ax[0][0] for first subplots


```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.plot(x,y)

# plt.show()
```

## Axes method

### Legend inferred from data

The easiest way to do this is to just give the lines labels, then call legend with default method.

The syntax is 

ax.legend(handles, labels, loc) # all the parameters can be omitted


```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "-b", label="sine") # note, legend has to have "label", otherwise, it won't show up
plt.plot(x, y2, "-r", label="cosine")
plt.legend(loc="upper left") # here, loc is used
```

### Custom Legend

```python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

legend_elements = [Line2D([0], [0], color='b', lw=4, label='Line'),
                   Line2D([0], [0], marker='o', color='w', label='Scatter',
                          markerfacecolor='g', markersize=15),
                   Patch(facecolor='orange', edgecolor='r',
                         label='Color Patch')]

# Create the figure
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.legend(handles=legend_elements, loc='center')
ax.plot(x, x**2,color='y', linestyle='dashed')
```

### Set twin axis

It is considered useful to have dual x or y axes in a figure. Moreso, when plotting curves with different units together. Matplotlib supports this with the twinxand twiny functions.

In the following example, the plot has dual y axes, one showing exp(x) and the other showing log(x)

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx() # key is here
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```

## Scatter

```python
np.random.seed(1)
```

```python
mu, sigma = 1, 2
scale = 0.1
size = 25
x = np.random.normal(mu, sigma, size=(size,))
e = np.random.normal(0, sigma*scale, size=(len(x),))

slope, intercept = 1.3, 0
fig, ax = plt.subplots(1,1, figsize=(6,6))
y = slope * x + intercept + e
ax.scatter(x, y, color='r', label="y vs x")
ax.set_title("linear regression")
ax.set_xlabel("X")
ax.set_ylabel("Y = slope*X + error")
ax.set_xlim(-2, 6)
ax.set_ylim(-3, 7)
ax.set_xticks([-2, 0, 2, 4, 6])
ax.set_yticks([-2, 0, 2, 4, 6])
ax.legend()
plt.show()
```

## Contour

- X, Y: These parameter are the coordinates of the values in Z.
- Z : This parameter is the height values over which the contour is drawn.
- levels : 
    - levelsint or array-like, optional
    - Determines the number and positions of the contour lines / regions.
    - If an int n, use MaxNLocator, which tries to automatically choose no more than n+1 "nice" contour levels between vmin and vmax.
    - If array-like, draw contour lines at the specified levels. The values must be in increasing order.
    - Suppose we have levels=[a], this is to draw Z=a

```python
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist) # note, X.shape = (100, 100), Y.shape = (100, 100)

Z = np.sqrt(X**2 + Y**2) # note Z.shape = (100, 100)
fig,ax=plt.subplots(1,3, figsize=(20, 5))
cp = ax[0].contourf(X, Y, Z)
fig.colorbar(cp, ax=ax[0]) # Add a colorbar to a plot
ax[0].set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax[0].set_ylabel('y (cm)')

# make the second plot, with levels
#Z = np.ones(len(X))
cp2 = ax[1].contour(X, Y, Z)

# if we only show one level
cp2 = ax[2].contour(X, Y, Z, levels=[1])

plt.show()
```

## one example


We will show in origianl coordinate $x^2 + y^2 = 1$ and how it is transformed after a linear matrix $A = [[1,1], [0,1]]$

```python
x = np.linspace(-1.0, 1.0, 100)

y_pos = np.sqrt(1- x**2)
y_neg = -np.sqrt(1- x**2)
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.plot(x, y_pos)
ax.plot(x, y_neg)
plt.show()
```

```python
def plot_2d_matrix(A):
    x = np.linspace(-1.0, 1.0, 100)

    y_pos = np.sqrt(1- x**2)
    y_neg = -np.sqrt(1- x**2)
    x_new_pos, y_new_pos = [], []
    for i in zip(x, y_pos):
        temp = np.array(i).reshape(2,1)
        new_vec = np.matmul(A, temp)
        #print(new_vec)
        x_new_pos.append(new_vec[0][0])
        y_new_pos.append(new_vec[1][0])
        #break

    x_new_neg, y_new_neg = [], []
    for i in zip(x, y_neg):
        temp = np.array(i).reshape(2,1)
        new_vec = np.matmul(A, temp)
        #print(new_vec)
        x_new_neg.append(new_vec[0][0])
        y_new_neg.append(new_vec[1][0])

    fig, ax = plt.subplots(1,1, figsize=(5,5))
    ax.plot(x_new_pos, y_new_pos)
    ax.plot(x_new_neg, y_new_neg)
    plt.show()
```
