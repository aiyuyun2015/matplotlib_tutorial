# Make quick plots

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
```

What plot does, is to use 2-d array, and plots 
$(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)$

Let's make the title and label

```python
plt.plot(x,y)

plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')

```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()
```

# Object-oriented interface

While it is easy to quickly generate plots with the matplotlib.pyplot module, the use of object-oriented approach is recommended as it gives more control and customization of your plots. Most of the functions are also available in the matplotlib.axes.Axes class.

The main idea behind using the more formal object-oriented method is to create figure objects and then just call methods or attributes off of that object. This approach helps better in dealing with a canvas that has multiple plots on it.

In object-oriented interface, Pyplot is used only for a few functions such as figure creation, and the user explicitly creates and keeps track of the figure and axes objects. At this level, the user uses Pyplot to create figures, and through those figures, one or more axes objects can be created. These axes objects are then used for most plotting actions.

To begin with, we create a figure instance which provides an empty canvas.

## How to make ax

	注意，一般都有两种方法建立axes
	fig = plt.figure()
	
	ax = fig.add_axes([x0, y0, width, height]) # method-1, create absolute axes, note, this will lead to issues such as not enough space for ticks, all the numbers are between [0, 1].
	
	ax = fig.add_subplot(rcn) # r--> row number, c --> column number, n --> current plot position
	
	ax = fig.add_subplot(111) # do you think this is equivalent to add_axes([0,0,1,1])? No, this is actually equal to [0.125,0.11,0.775,0.77], some spaces are reserved for ticks, labels, etc.
	
	The preferred method is:
	
	fig, ax = plt.subplots(1, 3) # ax will be an array, to access them, use ax[0], ax[1], ax[2]; if we have a 2d array, then we need do ax[0][0] for first subplots


```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.plot(x,y)

# plt.show()
```

## Make subplots use subplots method

```python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots(1,3)

ax[0].plot(x, x, label='line')
ax[0].set_title("fig1")
ax[1].plot(x, x**2)
ax[1].set_title("fig2")
ax[2].plot(x, x**3)
ax[2].set_title("fig3")
```

## Axes method

### Legend inferred from data

The easiest way to do this is to just give the lines labels, then call legend with default method.

The syntax is 

ax.legend(handles, labels, loc) # all the parameters can be omitted


```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "-b", label="sine")
plt.plot(x, y2, "-r", label="cosine")
plt.legend(loc="upper left") # here, loc is used
```

### Custom Legend

```python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

legend_elements = [Line2D([0], [0], color='b', lw=4, label='Line'),
                   Line2D([0], [0], marker='o', color='w', label='Scatter',
                          markerfacecolor='g', markersize=15),
                   Patch(facecolor='orange', edgecolor='r',
                         label='Color Patch')]

# Create the figure
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.legend(handles=legend_elements, loc='center')
ax.plot(x, x**2,color='y', linestyle='dashed')
```

### Line style method

Using **linestyle** Argument:

- Solid
- Dashed
- Dotted
- Dashdot
- None


### Color method

Use **color** argument
- 'b' as blue
- 'g' as green
- 'r' as red
- 'c' as cyan
- 'm' as magenta
- 'y' as yellow
- 'k' as black
- 'w' as white

### Marker method

use **marker** argument
- '.' point
- 'o' circle
- 's' square
- '+' plus

etc

```python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
# Create the figure
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, x**2, color='y',linestyle='dashed')
```

### Grid 

```python
#fig, axes = plt.subplots(1,3, figsize = (12,4)) # fig size default is (6.4, 4.8)
fig, axes = plt.subplots(1,3, figsize=(15, 5)) # fig size default is (6.4, 4.8) --> (width, height)

x = np.arange(1,11)
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)  # display
axes[0].set_title('default grid')
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls = '-.', lw = 0.25) # specify more details of the grid
axes[1].set_title('custom grid')
axes[2].plot(x,x)  # no grid! default is false
axes[2].set_title('no grid')
```
