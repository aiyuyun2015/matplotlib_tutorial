### 1 - Change the y-axis unit

```python
import matplotlib.ticker as ticker
```

```python
def format_func(value, scale=10000, y_axis_unit='bps'):
    """
    Convert y-axis to another unit, e.g., use bps.
    """
    
    # todo: why can't I use scale here??? have to hard-code this 10000
    scaled_value = value * 10000
    return f"{scaled_value:.1f} {y_axis_unit}"
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [-2e-4, 4e-4, 6e-4, 8e-4, 10e-4]

# Create a figure and axis
fig, ax = plt.subplots(1, 1, squeeze=False)

# Plot the data
ax[0][0].plot(x, y)

# Add labels and title
ax[0][0].set_xlabel('X-axis')
ax[0][0].set_ylabel('Y-axis')
ax[0][0].set_title('Sample Plot')

ax[0][0].yaxis.set_major_formatter(ticker.FuncFormatter(format_func))


# Display the plot
plt.show()
```

#### Legend is too big, move it outside of the plot

```python
# Generate some data for demonstration
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the curves
curve1, = ax.plot(x, y1, label='Curve 1')
curve2, = ax.plot(x, y2, label='Curve 2')

# Add the legend with customized properties
x, y = 0.6, 1.5
x, y = (0.6, -0.5)
legend = ax.legend(bbox_to_anchor=(x, y), title='Legend2')

# Modify the legend's frame properties
legend.get_frame().set_linewidth(3)
legend.get_frame().set_edgecolor('red')

# Show the plot
plt.show()
```
